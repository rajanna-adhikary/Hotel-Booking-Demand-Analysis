```python
import pandas as pd #data handling in tabular form
import numpy as np # for numerical operations
import matplotlib.pyplot as plt # for visualizations
import seaborn as sns # for better or advanced visualizations
from scipy import stats # for statistical tests

```

```python
file_path=r"C:\Users\RAJANNA ADHIKARY\Downloads\archive\hotel_bookings.csv"
df=pd.read_csv(file_path)
#for viewing the first 5 rowsssssss
df.head()
```

```python
#for the null values
#df.isnull().sum() this tells the null value in every row even the null value is 0 it will show up
# Checking missing values(only the null value rows will show up)
missing_values = df.isnull().sum()
missing_values[missing_values > 0]
```

```python
# data cleaning
# duplicates and also cleanimg
print(f"Duplicate Rows: {df.duplicated().sum()}")
df.drop_duplicates(inplace=True)
```

```python
#handling missing values
df.dropna(subset=['country', 'children'], inplace=True)  # Drop critical missing values
df.fillna({'agent': 0, 'company': 0}, inplace=True)  # Fill missing values with 0
```

```python
# feature engineering (creting new features)
df['total_nights'] = df['stays_in_weekend_nights'] + df['stays_in_week_nights']
df['is_cancelled'] = df['reservation_status'].apply(lambda x: 1 if x == 'Canceled' else 0)
```

```python
df[['stays_in_weekend_nights', 'stays_in_week_nights', 'total_nights', 'reservation_status', 'is_cancelled']].head()
#to checkt that our job is done or not
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
df['hotel'] = encoder.fit_transform(df['hotel'])
df['customer_type'] = encoder.fit_transform(df['customer_type'])
```

```python
# EDA and insights
sns.countplot(x='hotel', data=df, hue='hotel', palette='coolwarm', legend=False)
plt.title("Number of Bookings by Hotel Type")
plt.xlabel("Hotel Type")
plt.ylabel("Count")
plt.xticks(ticks=[0, 1], labels=['City Hotel', 'Resort Hotel'])
plt.show()
```

```python
#cancellation rate by hotels 
plt.figure(figsize=(8,5))
sns.barplot(x='hotel', y='is_cancelled', data=df, palette='coolwarm')
plt.title("Cancellation Rate by Hotel Type")
plt.xticks(ticks=[0,1], labels=['City Hotel', 'Resort Hotel'])
plt.show()
```

```python


# Count of cancellations for each hotel type
city_cancelled = df[df['hotel'] == 0]['is_canceled'].sum()
resort_cancelled = df[df['hotel'] == 1]['is_canceled'].sum()

# Total bookings for each hotel type
total_city = df[df['hotel'] == 0].shape[0]
total_resort = df[df['hotel'] == 1].shape[0]

# Perform Chi-Square Test
chi2, p, dof, expected = stats.chi2_contingency([[city_cancelled, total_city - city_cancelled], 
                                                 [resort_cancelled, total_resort - resort_cancelled]])
print(f"Chi-Square Test p-value: {p}")

if p < 0.05:
    print("Statistically significant: City hotels have higher cancellations.")
else:
    print("No significant difference in cancellation rates.")
```

```python


# Splitting data into two groups: Canceled & Not Canceled
lead_time_canceled = df[df['is_canceled'] == 1]['lead_time']
lead_time_not_canceled = df[df['is_canceled'] == 0]['lead_time']

# Perform Independent T-test
t_stat, p = stats.ttest_ind(lead_time_canceled, lead_time_not_canceled, equal_var=False)
print(f"T-Test p-value: {p}")

if p < 0.05:
    print("Statistically significant: Longer lead times tend to result in more cancellations.")
else:
    print("No significant correlation between lead time and cancellations.")
```

```python
from statsmodels.stats.proportion import proportions_ztest

# Count of families with children in each hotel type
resort_with_children = df[(df['hotel'] == 1) & (df['children'] > 0)].shape[0]
city_with_children = df[(df['hotel'] == 0) & (df['children'] > 0)].shape[0]

# Total bookings for each hotel type
total_resort = df[df['hotel'] == 1].shape[0]
total_city = df[df['hotel'] == 0].shape[0]

# Perform Proportion Z-Test
prop_test = proportions_ztest([resort_with_children, city_with_children], [total_resort, total_city])
print(f"Proportion Test p-value: {prop_test[1]}")

if prop_test[1] < 0.05:
    print("Statistically significant: Families with children prefer resort hotels.")
else:
    print("No significant difference in preference for city vs. resort hotels among families.")
```
